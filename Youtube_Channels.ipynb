```python

data2['Video Uploads'] = pd.to_numeric(data2['Video Uploads'], errors='coerce')

# Convert 'Subscribers' column to numeric, replacing non-numeric values with NaN
data2['Subscribers'] = pd.to_numeric(data2['Subscribers'], errors='coerce')

# Fill NaN values with 0 and convert to integers
data2['Video Uploads'] = data2['Video Uploads'].fillna(0).astype(int)
data2['Subscribers'] = data2['Subscribers'].fillna(0).astype(int)
```
